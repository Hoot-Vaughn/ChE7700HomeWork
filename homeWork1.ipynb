```python
#    3. A prime number is a whole number greater than 1 whose only factors are 1 and itself. A factor is a whole number that can be divided evenly into another number. 
# (a) Based on the definition of prime number, define a function to determine whether an input integer is a prime number or not
# (b) Find all prime numbers below 100
# (c) Define a function to find all prime numbers below a custom-defined number, n
# (d) Call your defined function to find all prime numbers with n = 1000

# 3a

def prime_num(x):

    if x<=1:
        print("Negative numbers are not prime numbers")
    for i in range(2, x):
        if x % i == 0:
            return 0
    return 1





```

```python
#     5. Solve following set of equations using linear algebra and numpy.

# 3a + 4b -5c = 10
# 5b + c = 8
# a + 3b + 6c = -5

x = np.array([[3,4,-5], [0,5,1], [1,3,6]])
y = np.array([10,8,-5])

z = np.linalg.solve(x,y)

a = z[0]
b = z[1]
c = z[2]
print(z)
print("a = ", a)
print("b = ", b)
print("c = ", c)
```

```python
#     6. Given an expression:

# y = 0.1x^3-x^2+5

# (a) Plot the function in the range of x from -5 to 5
# (b) Find the maximum and minimum value of this function and add them with points and labels on the same figure
r1 =-5
r2=5
X=[]
Y=[]

def yFunction(x):

    y = 0.1*x**3-x**2+5
    return(y)

ii = 0

for i in range(r1,r2):
    x = ii
    X.append(x)
    y = yFunction(x)
    Y.append(y)
    ii = ii+1


plt.figure(figsize=(10, 6))
plt.plot(X, Y, label='y = 0.1x^3 - x^2 + 5', color='blue')

ymin = np.min(Y)
xmin = X[np.argmin(Y)]
ymax = np.max(Y)
xmax = X[np.argmax(Y)]

plt.scatter(xmin, ymin, color='red', zorder=5)
plt.annotate(f'Min: ({xmin:.2f}, {ymin:.2f})', xy=(xmin, ymin), xytext=(-50, -30), textcoords='offset points')


plt.scatter(xmax, ymax, color='green', zorder=5)
plt.annotate(f'Max: ({xmax:.2f}, {ymax:.2f})', xy=(xmax, ymax), xytext=(20, 20), textcoords='offset points')



plt.xlabel('x values')
plt.ylabel('y values')
plt.legend()

plt.show()
```
